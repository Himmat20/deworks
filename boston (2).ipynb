```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = workface.iloc[:,0:13] 
y = workface.iloc[:,-1] 
```

```python
y = np.round(workface['Price'])
bestfeatures = SelectKBest(score_func=chi2, k=5)
fit = bestfeatures.fit(X,y)
workfacescores = pd.DataFrame(fit.scores_)
workfacecolumns = pd.DataFrame(X.columns)
featureScores = pd.concat([workfacecolumns,workfacescores],axis=1)
featureScores.columns = ['Specs','Score'] 
featureScores
```

```python
print(featureScores.nlargest(5,'Score')) 
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_) 
```

```python

feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
x = workface.iloc[:,0:13]
y = workface.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_train)
```

```python
print("Training accuracy :" , model.score(x_train,y_train)*100)
```

```python
print("Testing accuracy:" ,model.score(x_test,y_test)*100)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print("model accuracy:" ,r2_score(y,model.predict(x))*100)
```

```python
plt.scatter(y_train,y_pred)
plt.xlabel('prices')
plt.ylabel('predicted prices')
plt.title('prices vs predicted prices')
plt.show()
```

```python
plt.scatter(y_pred,y_train-y_pred)
plt.title('predicted vs residuals')
plt.xlabel("predicted")
plt.ylabel('residuals')
plt.show()
```

```python
sns.distplot(y_train-y_pred)
plt.title("Histogram of residuals")
plt.xlabel('residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
x=workface.iloc[:,[-1,5,10,4,9]]
y=workface.iloc[:,[-1]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(x_train,y_train)
```
