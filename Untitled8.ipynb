```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
ticker = "AAPL"  
data = yf.download(ticker, start="2015-01-01", end="2024-01-01")
```

```python
data = data[['Close']]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data, label=f"{ticker} Closing Prices")
plt.title(f"{ticker} Closing Price History")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)
```

```python
sequence_length = 60  
X, y = [], []

for i in range(sequence_length, len(data_scaled)):
    X.append(data_scaled[i-sequence_length:i, 0])  
    y.append(data_scaled[i, 0])  

X, y = np.array(X), np.array(y)
```

```python
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python

model = Sequential([
    SimpleRNN(units=50, activation='tanh', return_sequences=True, input_shape=(X.shape[1], 1)),
    SimpleRNN(units=50, activation='tanh', return_sequences=False),
    Dense(units=1) 
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, label="Actual Prices", color="blue")
plt.plot(y_pred, label="Predicted Prices", color="red")
plt.title(f"{ticker} Price Prediction")
plt.xlabel("Time")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```
